```python
import numpy as np
# 1. Create a numpy array of size 2 x 2
array_2x2 = np.zeros((2, 2), dtype=int)
print(f"1. 2x2 Array:\n{array_2x2}\n")

# 2. Create a random numpy array of size 3 x 4
random_3x4 = np.random.rand(3, 4)
print(f"2. Random 3x4 Array:\n{random_3x4}\n")

# 3. Create a numpy array of ones and a numpy array of zeroes
print("3. Array of Ones and Array of Zeroes")
ones_array = np.ones((2, 3), dtype=int)
print(f"Ones Array (2x3):\n{ones_array}")
zeros_array = np.zeros((4, 1), dtype=int)
print(f"Zeroes Array (4x1):\n{zeros_array}\n")

# 4. Create a random numpy array of size 6x3 and transpose it
print("4. Random 6x3 Array and Transpose")
original_6x3 = np.random.randint(10, size=(6, 3))
print(f"Original 6x3 Array:\n{original_6x3}")
transposed_array = original_6x3.T
print(f"Transposed 3x6 Array:\n{transposed_array}\n")


# 5. Create two random numpy arrays of size n x p and p x m and then compute the dot product
print("5. Dot Product of n x p and p x m Arrays")
n = 3
p = 2
m = 4
A = np.random.randint(1, 5, size=(n, p))
print(f"Array A ({n}x{p}):\n{A}")
B = np.random.randint(1, 5, size=(p, m))
print(f"Array B ({p}x{m}):\n{B}")
dot_product = A @ B
print(f"Dot Product A @ B ({n}x{m}):\n{dot_product}")
```
